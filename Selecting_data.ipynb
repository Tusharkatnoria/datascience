# selecting data using loc and iloc

* using 'loc' to get through the row name or the column name
* but 'iloc' is used to get through its position according to the index


```python
import numpy as np
import pandas as pd

df= pd.read_csv('laptop_price.csv' , encoding = 'latin-1')
df.head()
```

```python

df.index
```

```python
df.set_index('laptop_ID').head()
```

```python
df.loc[1,['Company','Product']] # to get the specific raw and the column
```

```python
# Selects all rows and only the 'Company' and 'Product' columns from the dataframe
df.loc[:, ['Company', 'Product']]
```

# to drop a row

```python
df.drop(2 ,axis=0,inplace=True)
df.drop(index=[3])
df.drop([]) # in that list we can drop the list of the index
```
