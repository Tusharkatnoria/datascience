```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptop_price.csv' , encoding= 'latin-1')
df.head()
```

```python
df[(df['Company'] == 'Apple') & (df['Price_euros'] > 2000)]
```

```python
df[(df['Company'] == 'Apple') | (df['Price_euros'] > 2000)]
```

```python
# creating more than two conditions using ...... NP SELECT STATEMENTS
# format_:

# conditions=[]
# values =[]   values to the corresponding solutions
```

```python
conditions = [ df['Price_euros']<2000,
               (df['Price_euros']>2000) & (df['Price_euros']<1000),
                (df['Price_euros']>2000)
                  ]
values=['cheaper','okok','expensive']
np.select(conditions, values,default='unknown')
```
